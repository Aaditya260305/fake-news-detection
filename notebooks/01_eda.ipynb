# EDA — Real or Fake + Fake News Detection

Reproduces the dataset statistics in paper Sec. V-A-3 and V-A-4 for the **English** subset of EKNet.

```python
import sys
sys.path.insert(0, '..')
from src.config import load_config
from src.data import load_real_or_fake, load_fake_news_detection, articles_to_dataframe
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns

cfg = load_config('../configs/default.yaml')
pd.set_option('display.max_colwidth', 120)
```

```python
rof = load_real_or_fake(cfg)
fnd = load_fake_news_detection(cfg)
print(f'Real-or-Fake : {len(rof):>5} articles')
print(f'Fake-News-Det: {len(fnd):>5} articles')
```

```python
df_rof = articles_to_dataframe(rof)
df_fnd = articles_to_dataframe(fnd)

fig, axes = plt.subplots(1, 2, figsize=(10, 3))
df_rof['label'].value_counts().plot.bar(ax=axes[0], title='Real-or-Fake (1=fake)')
df_fnd['label'].value_counts().plot.bar(ax=axes[1], title='Fake-News-Detection (1=fake)')
for ax in axes: ax.set_xlabel(''); ax.set_ylabel('count')
plt.tight_layout()
```

```python
df_rof['len_chars'] = df_rof['text'].str.len()
df_fnd['len_chars'] = df_fnd['text'].str.len()
print('Real-or-Fake  mean chars:', df_rof['len_chars'].mean())
print('Fake-News-Det mean chars:', df_fnd['len_chars'].mean())

fig, ax = plt.subplots(figsize=(8, 3))
sns.kdeplot(df_rof[df_rof['len_chars'] < 20000]['len_chars'], label='Real-or-Fake', ax=ax)
sns.kdeplot(df_fnd[df_fnd['len_chars'] < 20000]['len_chars'], label='Fake-News-Det', ax=ax)
ax.set_xlabel('article length (chars)'); ax.legend()
```
